#### What is the trend of gold prices over time?

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'],df['Price Today'])
plt.title('Trend of Gold Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Gold Price')
plt.show()
```

#### What are the monthly and yearly average gold prices?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
monthly_avg_price = df.groupby(['Year','Month'])['Price Today'].mean().unstack()
monthly_avg_price
```

```python
year_avg_price = df.groupby('Year')['Price Today'].mean()
year_avg_price
```

#### What is the moving average of gold prices over a 7-day window?

```python
df['7-day-moving-avg'] = df['Price Today'].rolling(window=7).mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'],df['Price Today'],label='Gold-Price')
plt.plot(df['Date'],df['7-day-moving-avg'],label='7-Day Moving Average', color='orange')
plt.title('Gold Price and 7-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Gold Price')
plt.legend()
plt.show()
```

####  How does the gold price volatility change over time?

```python
df['Daily_Return'] = df['Price Today'].pct_change()
df['Volatility'] = df['Daily_Return'].rolling(window=30).std() * np.sqrt(30)

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Volatility'])
plt.title('Volatility of Gold Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.show()
```

#### What is the percentage change in gold prices from day-to-day?

```python
df['Daily_Percentage_Change'] = df['Price Today'].pct_change() * 100

df[['Date', 'Daily_Percentage_Change']].head()
```

####  What is the correlation between gold prices and other economic metrics provided in the dataset?

```python
correlation_metrices = df.corr()
plt.figure(figsize=(18,16))
sns.heatmap(correlation_metrices,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#### Is there a significant relationship between gold prices and inflation rates?

```python
correlation_inf_price = df[['Price Today','Monthly Inflation Rate']].corr()
correlation_inf_price
```

#### How do gold prices correlate with Crude Price, if available?

```python
correlation_currency = df[['Price Today', 'Crude']].corr()
correlation_currency
```

#### What are the annual trends in gold prices using a line graph?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Price Today')
plt.title('Annual Trends in Gold Prices')
plt.xlabel('Year')
plt.ylabel('Gold Price')
plt.show()
```

####  How does the gold price distribution vary across different months of the year?

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Month', y='Price Today')
plt.title('Gold Price Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Gold Price')
plt.show()
```

#### Can we identify any seasonal patterns in gold prices?

```python
seasonal_pattern = df.groupby(['Month'])['Price Today'].mean()

plt.figure(figsize=(12, 6))
seasonal_pattern.plot(kind='bar')
plt.title('Seasonal Patterns in Gold Prices')
plt.xlabel('Month')
plt.ylabel('Average Gold Price')
plt.show()
```
